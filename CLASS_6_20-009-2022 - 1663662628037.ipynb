###  Handling Outliers:

**Outlier:**
        
        The records which lie outside the acceptable range, where the cutoff can be calculated using any of the detection methods.
        
* Detection:

    Detection can be using the statistical functions:
            i. %tile  method: commonly  used, its reasy to work with, no calculation involved
            ii. IQR method: lot of calculations are required
            iii. STD. method

* Treatment:
        
        Remove the outliers using pd.clip(lower = lc, upper =  uc)

#### %tile Method:

        Select the 1% of the records from both the ends as an outlier.
        lc = p1
        uc = p99
        
#### IQR method: Inter Quartile Range

        IQR =  P75 - P25  or Q3 - Q1
        
        p0(min), p100(max), p50, p25, p75 -----> .quantile(q = 0.5, 0.25, 0.75)
        
        lc =  p25 - 1.5 * IQR
        uc =  p75 + 1.5 * IQR
        
#### STD. method:

        lc =  mean - 2.5 * std
        uc =  mean + 2.5 * std
